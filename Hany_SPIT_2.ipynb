```python
#Let's use the popular Apriori Algorithm on df1 now for France.
dfx = (df1[df1['Country'] == 'France'])
df3 = dfx.pivot_table(index='TransactionId', columns='ItemCode', values='Count', aggfunc=np.sum, fill_value = 0)
```

```python
print(df3)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
df3 = df3.applymap(encode_units)
df3
```

```python
frequent_itemsets1 = apriori(df3, min_support=0.08, use_colnames=True)
rules1 = association_rules(frequent_itemsets1, metric="lift", min_threshold=1.1)
```

```python
rules1.head(20)
```

```python
#Let's use the popular Apriori Algorithm on df2 now.
dfx = (df2[df2['Country'] == 'France'])
df5 = dfx.pivot_table(index='TransactionId', columns='ItemCode', values='Count', aggfunc=np.sum, fill_value = 0)
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
df5 = df5.applymap(encode_units)
frequent_itemsets2 = apriori(df5, min_support=0.08, use_colnames=True)
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1.1)
```

```python
rules2.head(20)
```

```python
rules2[ (rules2['lift'] >= 5) & (rules2['confidence'] >= 0.7) ]
```

```python
rules2.to_csv('France.csv')
```
